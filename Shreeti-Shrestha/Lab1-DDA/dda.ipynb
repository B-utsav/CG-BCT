```python
import matplotlib.pyplot as plt
x1= int(input("enter value of x1"))
x2= int(input("enter value of x2"))
y1= int(input("enter value of y1"))
y2= int(input("enter value of y2"))
def function_dda():
    diffx = x2-x1
    diffy = y2-y1
    steps = max(abs(diffx),abs(diffy))
    xin = diffx/steps
    yin = diffy/steps
    x = x1
    y = y1
    plotx =[x1]
    ploty =[y1]
    for i in range (steps):
      
        
        x = (x + xin)
        y = (y + yin)
        plotx.append(x)
        ploty.append(y)
        print(f'({x},{y})')
    plt.plot(plotx,ploty,marker ='^',color = 'maroon')
    plt.show()
function_dda()
```
